```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
Customer_info_df = pd.read_csv("D:\下载\Customer_info_df.csv")
Customer_info_df
```

```python
Customer_info_df.dropna(axis=0, how='any', inplace=True)
Customer_info_df.info()
```

```python
Customer_info_df['DOB'] = pd.to_datetime(Customer_info_df['DOB'])
Customer_info_df.info()
```

```python
#Customer_info_df['DOB'].dt.year
Customer_info_df = Customer_info_df.drop(Customer_info_df[Customer_info_df['DOB'].dt.year < 1900].index)
Customer_info_df.sort_values(by='DOB', ascending=True, inplace=False)
```

```python
Customer_info_df.drop('country', axis=1, inplace=True)
Customer_info_df
```

```python
Customer_info_df['DOB'] = Customer_info_df.DOB.apply(lambda x: x.year)
Customer_info_df
```

```python
sns.histplot(Customer_info_df['list_price'], bins=20, kde=True)
plt.title('Distribution of List Price')
plt.show()
```

```python
sns.scatterplot(x='age', y='past_3_years_bike_related_purchases', data=Customer_info_df)
plt.title('Age vs Bike Related Purchases')# age group
plt.show()
```

```python
sns.countplot(x='gender', data=Customer_info_df)
plt.title('Count of Customers by Gender')
plt.show()
```

```python
Customer_info_df['transaction_date'] = pd.to_datetime(Customer_info_df['transaction_date'])
Customer_info_df.set_index('transaction_date', inplace=True)
Customer_info_df['list_price'].resample('M').sum().plot(title='Monthly Sales Trend')
plt.show()
```

```python
correlation_matrix = Customer_info_df[['list_price', 'standard_cost', 'past_3_years_bike_related_purchases', 'age', 'tenure', 'property_valuation']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
Customer_info_df.reset_index(inplace=True)
Customer_info_df
```

```python
Customer_info_df['transaction_date'] = pd.to_datetime(Customer_info_df['transaction_date'])
Customer_info_df['transaction_month'] = Customer_info_df['transaction_date'].dt.to_period('M')
Customer_info_df['cohort'] = Customer_info_df.groupby('customer_id')['transaction_month'].transform('min')
Customer_info_df
```

```python
cohort_data = Customer_info_df.groupby(['cohort', 'transaction_month']).agg(avg_order_value=('list_price', 'mean')).reset_index()
cohort_data
```

```python
cohort_matrix = cohort_data.pivot(index='cohort', columns='transaction_month', values='avg_order_value')
plt.figure(figsize=(12, 8))
sns.heatmap(cohort_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Cohort Analysis - Average Order Value')
plt.show()
```
